```python
import os
import pandas as pd
import pyarrow.parquet as pq

def read_and_process_parquet_files(directory):
    # List all Parquet files in the directory
    parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

    # Read and concatenate all Parquet files into a single DataFrame
    df_list = [pd.read_parquet(file) for file in parquet_files]
    combined_df = pd.concat(df_list, ignore_index=True)

    # Remove duplicate rows
    combined_df.drop_duplicates(inplace=True)

    return combined_df

# Example usage
dir1 = r"D:\RISHIN\TESTING\TEST_3\ILC2024_EUWS_PLA_WI_EP_BE_EUR_Losses\EP\Admin1\GU"
dir2 = r"Y:\PROJECT\Renu\ILC_2024\Testing_EUWS\Results\ILC\BE\ILC2024_EUWS_PLA_WI_EP_BE_EUR_Losses\EP\Admin1\GU"

df1 = read_and_process_parquet_files(dir1)
df2 = read_and_process_parquet_files(dir2)

print("DataFrame from directory 1:")
print(df1)

print("\nDataFrame from directory 2:")
print(df2)
```
