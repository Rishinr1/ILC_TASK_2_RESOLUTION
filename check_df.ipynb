```python
import os
import pandas as pd
import pyarrow.parquet as pq

def read_and_process_parquet_files(directory):
    # List all Parquet files in the directory
    parquet_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.parquet')]

    # Read and concatenate all Parquet files into a single DataFrame
    df_list = [pd.read_parquet(file) for file in parquet_files]
    combined_df = pd.concat(df_list, ignore_index=True)

    # Remove duplicate rows
    combined_df.drop_duplicates(inplace=True)

    return combined_df

# Example usage
dir1 = r"D:\RISHIN\TESTING\TEST_9\ILC2024_EUWS_EP_PLA_EUR_Losses\EP\Cresta_Lob\GR"
dir2 = r"Y:\PROJECT\Renu\ILC_2024\Testing_EUWS\Results\ILC\BE\ILC2024_EUWS_PLA_WI_EP_BE_EUR_Losses\EP\Cresta_Lob\GR"

df3 = read_and_process_parquet_files(dir1)
df4 = read_and_process_parquet_files(dir2)
```

```python
import os
import pyarrow.parquet as pq

def check_files_in_directory(directory_path):
    # List all files in the directory
    files = [f for f in os.listdir(directory_path) if f.endswith('.parquet')]
    
    # Iterate through each file and check the number of rows
    for file in files:
        file_path = os.path.join(directory_path, file)
        table = pq.read_table(file_path)
        if table.num_rows != 48:
            print(f"File {file} does not have 48 rows, it has {table.num_rows} rows.")

# Directory path
directory_path = r"D:\RISHIN\TESTING\TEST_9\ILC2024_NZFL_EP_PLA_NZD_Losses\EP\Admin1_Lob\GU"

# Check files in the directory
check_files_in_directory(directory_path)
```
