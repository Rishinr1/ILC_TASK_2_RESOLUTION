```python
import os
import pyarrow.parquet as pq
import pyarrow as pa

def read_parquet_files(directory):
    """Read all Parquet files from a directory and combine them into a single table."""
    tables = []
    for filename in os.listdir(directory):
        if filename.endswith('.parquet'):
            file_path = os.path.join(directory, filename)
            table = pq.read_table(file_path)
            tables.append(table)
    combined_table = pa.concat_tables(tables)
    return combined_table

def print_distinct_weights(directory):
    """Print all distinct values in the 'Weight' column from Parquet files in the directory."""
    combined_table = read_parquet_files(directory)
    distinct_weights = combined_table['Weight'].unique()
    print("Distinct weights:", distinct_weights)

# Example usage
directory = r'D:\RISHIN\14_2_1ILC_NZFL\PLT\Risk_Lob\GU\PeriodRange=1-250000'
print_distinct_weights(directory)
```

```python
def concatenate_parquet_files(directory_path):
    # List all Parquet files in the directory
    files = [f for f in os.listdir(directory_path) if f.endswith('.parquet')]
    
    # Read each Parquet file into a PyArrow table
    tables = [pq.read_table(os.path.join(directory_path, f)) for f in files]
    
    # Concatenate all tables into a single table
    concatenated_table = pa.concat_tables(tables)
    
    return concatenated_table

# Directory path
directory_path = r'D:\RISHIN\TESTING\TEST_15\ILC2024_NZFL_EP_PLA_NZD_Losses\EP\Lob\GR'
directory_path2 = r'D:\RISHIN\TESTING\TEST_15\ILC2024_NZFL_EP_PLA_NZD_Losses\EP\Lob\GR'


# Concatenate Parquet files and read the result
concatenated_table = concatenate_parquet_files(directory_path)
concatenated_table_2 = concatenate_parquet_files(directory_path2)

df3 = concatenated_table.to_pandas()
df4 = concatenated_table_2.to_pandas()


```

```python
import os
import pyarrow.parquet as pq

def check_files_in_directory(directory_path):
    # List all files in the directory
    files = [f for f in os.listdir(directory_path) if f.endswith('.parquet')]
    
    # Iterate through each file and check the number of rows
    for file in files:
        file_path = os.path.join(directory_path, file)
        table = pq.read_table(file_path)
        if table.num_rows != 48:
            print(f"File {file} does not have 48 rows, it has {table.num_rows} rows.")

# Directory path
directory_path = r"D:\RISHIN\TESTING\TEST_9\ILC2024_NZFL_EP_PLA_NZD_Losses\EP\Admin1_Lob\GU"

# Check files in the directory
check_files_in_directory(directory_path)
```
